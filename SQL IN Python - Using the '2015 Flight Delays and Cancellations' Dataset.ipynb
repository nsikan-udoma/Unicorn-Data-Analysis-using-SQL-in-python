```python
query = """
SELECT 
    f.origin_airport AS origin_airport,
    ap.airport AS airport_name,
    ap.city AS city,
    ap.state AS state,
    SUM(f.cancelled) AS no_of_cancellations
FROM flights AS f
JOIN airlines AS al
    ON f.airline = al.IATA_CODE
JOIN airports AS ap
    ON f.origin_airport = ap.IATA_CODE
WHERE f.cancelled = 1
GROUP BY 1
ORDER BY 5 DESC
"""
sqldf(query)
```

```python
query = """
SELECT 
    f.origin_airport AS origin_airport,
    ap.airport AS airport_name,
    ap.city AS city,
    ap.state AS state,
    COUNT(f.departure_delay) AS no_of_delayed_flights
FROM flights AS f
JOIN airlines AS al
    ON f.airline = al.IATA_CODE
JOIN airports AS ap
    ON f.origin_airport = ap.IATA_CODE
WHERE f.departure_delay > 0
GROUP BY 1
ORDER BY 5 DESC
"""
sqldf(query)
```

```python
query = """
SELECT 
    f.origin_airport AS origin_airport,
    ap.airport AS airport_name,
    ap.city AS city,
    ap.state AS state,
    SUM(f.departure_delay) AS total_delay_time_mins
FROM flights AS f
JOIN airlines AS al
    ON f.airline = al.IATA_CODE
JOIN airports AS ap
    ON f.origin_airport = ap.IATA_CODE
WHERE f.departure_delay > 0
    AND f.cancelled != 1
GROUP BY 1
ORDER BY 5 DESC
"""
sqldf(query)
```

```python
query = """
SELECT 
    f.origin_airport AS origin_airport,
    ap.airport AS airport_name,
    ap.city AS city,
    ap.state AS state,
    COUNT(f.departure_delay) AS count
FROM flights AS f
JOIN airlines AS al
    ON f.airline = al.IATA_CODE
JOIN airports AS ap
    ON f.origin_airport = ap.IATA_CODE
WHERE f.departure_delay > 0
    AND f.cancelled = 1
GROUP BY 1
ORDER BY 5 DESC
"""
sqldf(query)
```

```python
query = """
SELECT 
    f.flight_number,
    f.origin_airport AS origin_airport,
    ap.airport AS airport_name,
    ap.city AS city,
    ap.state AS state, 
    f.cancelled
FROM flights AS f
JOIN airlines AS al
    ON f.airline = al.IATA_CODE
JOIN airports AS ap
    ON f.origin_airport = ap.IATA_CODE
WHERE f.departure_delay > 0
    AND f.cancelled = 1
    AND f.origin_airport = 'ORD'
"""
sqldf(query)
```
